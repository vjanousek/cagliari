#### Example 2: Simple diagrams


```r
loadData("data/dolerites_test.data",sep="\t")
# Binary plot
options(repr.plot.width=7, repr.plot.height=5)
#binary(x="SiO2",y="Y",ymin=0,ymax=50,main="My Binary plot")
figCex(1.8)

# A single classification diagram, zoom in
plotDiagram("LarochePlut",F,F)
figXlim(c(1000,2700))
figYlim(c(1000,2500))
figRemove()
```

```r
#loadData("data/dolerites_test.data",sep="\t")
# Binary plot
options(repr.plot.width=5, repr.plot.height=5)
#binary(x="SiO2",y="Y",ymin=0,ymax=50,main="My Binary plot")
figCex(1.8)

# A single classification diagram, zoom in
plotDiagram("LarochePlut",F,F)
figXlim(c(1000,2700))
figYlim(c(1000,2500))
figRemove()

# Harker plot (a simple plate)
multipleMjr("SiO2")
plateCex(1.7)
plateCexLab(1.2)
plateXLim(c(45,60))
plateBW()
```

### Example 3: Custom plates

```r
loadData("data/dolerites_test.data",sep="\t")

# Custom plate
multiplePerPage(4,nrow=2,ncol=2,dummy=F,title="Geotectonic")
plotDiagram("Meschede",F,F,main=" ")
plateExtract("Wood",3,main=" ")
plotDiagram("PearceNbThYb",F,F,main=" ")
plotDiagram("PearceNbTiYb",F,F,main=" ")
plateLabelSlots(text=letters,cex=1.5,pos="topleft")
plateCex(1.7)
plateCexLab(1.2)
```

### Example 4: Spiderplots

```r
loadData("data/dolerites_test.data",sep="\t")
groupsByLabel("region")

# REE spider
spider(WR,"Boynton",1,100,pch="*",col="blue",cex=2)
spider.contour("Boynton","MgO","jet.colors",ymin=1,ymax=100,pch=15)

# NMORB Sun and McDonough
spider(WR,"^NMORB..Sun",0.1,500,offset=TRUE)
figMulti(shaded.col="khaki",plot.symb=FALSE,nrow=2,ncol=2)
plateLabelSlots(text=letters,cex=1.5,pos="topright")
plateCex(2) 
plateCexLab(2)
```

### Example 5: Dataset switching

```r
# Loading and storing the 1st dataset
loadData("data/sazava.data",sep="\t")
pokeDataset("sazava")
# Second
loadData("data/dolerites_test.data",sep="\t")

options(repr.plot.width=5, repr.plot.height=5)
# Overplotting classification diagram
plotDiagram("TAS",F,F)
figCol("gray")
overplotDataset("sazava",pch="*",col="red",cex=2)
# Switching back to the dolerites dataset
peekDataset("dolerites_test.data")
```

```r
# Loading and storing the 1st dataset
loadData("data/sazava.data",sep="\t")
pokeDataset("sazava")
# Second
loadData("data/dolerites_test.data",sep="\t")

# Overplotting on spiderplot
#spider(WR,"Boynton",1,1000,cex=0,join=TRUE,offset=TRUE,centered=FALSE,xrotate=FALSE,xaxs="r")
#figCol("darkblue")
#overplotDataset("sazava")

# Switching back to the dolerites dataset
# Dolerites as grey field
spider(WR,"Boynton",0.1,1000,field=TRUE,fill.col=TRUE,shaded.col="khaki") 
overplotDataset("sazava")

# Getting rid of the datasets other than the current
# (dolerites)
purgeDatasets()
```
